### to check for similar users in the album file

```python
art_id=16
similar_users= album_data[(album_data["ArtistId"]== art_id)]['AlbumId'].unique()
```

```python
similar_users
```

```python
def find_similar_artistid(art_id):
    similar_users= album_data[(album_data["ArtistId"]== art_id)]['AlbumId'].unique()
    similar_user_recs = album_data[(album_data["AlbumId"].isin(similar_users))]
    similar_user_recs = similar_user_recs.value_counts() / len(similar_users)
    rec_percentages = pd.concat([similar_user_recs], axis=1)
    print(rec_percentages)
```

```python
for i in range(len(artist_data)):
    find_similar_artistid(artist_data['artistId'][i])
```
